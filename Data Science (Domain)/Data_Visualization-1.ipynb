```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10)
plt.plot(x, np.sin(x))
```

```python
x = np.linspace(0,10)
plt.plot(x, np.sin(x), color = 'red')
```

```python
x = np.linspace(0,10)
plt.plot(x, np.sin(x), color = 'red', linestyle = 'dotted')
```

```python

plt.plot(x, np.sin(x), linestyle = 'dashed')
```

```python
x = np.linspace(0,10)
plt.plot(x, np.sin(x), color = 'red', linestyle = 'solid')
```

```python
x = np.linspace(0,10)
plt.plot(x, np.sin(x), linestyle = 'dashdot')
```

```python
plt.plot(x, np.cos(x), 'y*')
plt.show()
```

```python
plt.plot(x, np.cos(x), 'yd')
plt.show()
```

```python
plt.plot(x, np.cos(x), 'ys')
plt.show()
```

```python
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
plt.bar(langs, students, width = 0.75, color = 'pink')
```

```python
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
plt.barh(langs, students, height = 0.75, color = 'blue')
```
